The sequence of triangle numbers is generated by adding the natural numbers. So the $7$<sup>th</sup> triangle number would be $1 + 2 + 3 + 4 + 5 + 6 + 7 = 28$. The first ten terms would be:
$$1, 3, 6, 10, 15, 21, 28, 36, 45, 55, \dots$$
Let us list the factors of the first seven triangle numbers:
\begin{align}
\mathbf 1 \colon 1\\
\mathbf 3 \colon 1,3\\
\mathbf 6 \colon 1,2,3,6\\
\mathbf{10} \colon 1,2,5,10\\
\mathbf{15} \colon 1,3,5,15\\
\mathbf{21} \colon 1,3,7,21\\
\mathbf{28} \colon 1,2,4,7,14,28
\end{align}
We can see that $28$ is the first triangle number to have over five divisors.
What is the value of the first triangle number to have over five hundred divisors?

```python
import math
import itertools
import collections
from time import perf_counter
```

```python
def triangle_generator():

    num = 0
    for curr in itertools.count(1):
        num += curr
        yield num
```

```python
def primes_gen():
    _primos = set()

    for valor in itertools.count(2):
        for p in _primos:
            if valor % p == 0:
                break
        else:
            _primos.add(valor)
            yield valor
```

```python
def necessary_primes(num: int):
    global primes, pr_gen
    while num > max(primes):
        primes.append(next(pr_gen))
```

```python
def prime_factors(num: int) -> dict:
    '''Counts how many times each prime is a factor of num and returns a dictionary of primes and each respective multiplicity.
    '''
    global primes
    factors = collections.defaultdict(int) # inicializa chaves inexistentes com 0
    if max(primes) < (num ** 0.5):
        necessary_primes(num)
        primes.sort()

    for prime in primes:
        while num % prime == 0:
            factors[prime] += 1
            num = num / prime
        if num == 1:
            break

    return factors
```

```python
def number_factors(factors: dict) -> int:
    result = 1
    for v in factors.values():
        result *= (v + 1)
    return result
```
