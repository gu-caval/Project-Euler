We shall say that an **n**-digit number is pandigital if it makes use of all the digits **1** to **n** exactly once. For example, **2143** is a **4**-digit pandigital and is also prime.<br><br>
What is the largest **n**-digit pandigital prime that exists?

```python
def pandigital_test(numbers: list) -> bool:
    '''Receives a list of integers and checks if they are colectively 1 to n Pan Digital
    (if all digits from 1 to n are present once and only once), where n is the length of
    list.
    '''
    test = True
    for i in range(1,len(numbers)+1):
        if i not in numbers:
            test = False
            break
    strings_set = {str(i) for i in numbers}
    size = len(numbers)
    return test and len(strings_set) == size #and '0' not in strings_set
```

```python
def primes_gen(max=None):
    if max < 2:
        return None
    
    primos = set()
    primos.add(2)
    
    if max == 2:
        return 2
    else:
        yield 2
    

    valor = 3
    while valor <= max:
        validador = True
        for p in primos:
            if valor%p == 0:
                validador = False
                break

        if validador:
            primos.add(valor)
            yield valor
            
        valor += 2
```

```python
result = 2
for prime in primes_gen(1000000000):
    string_list = [int(i) for i in str(prime)]
    if pandigital_test(string_list):
        result = prime
        print(f'{prime} is the largest so far.')
```

```python
from time import perf_counter

def primes_from_largest_down (max: int):
    # if type(max) not int:
    #     raise TypeError('argument is not an integer')
    # Initializes sieve
    start = perf_counter()
    sieve = [0] * (max + 2)
    # Marks all even numbers > 2 as not primes
    for n in range(4, max+1, 2):
        sieve[n] = 1
    # Tests the numbers not marked as not primes, what is left is the primes at the spots with 0
    for n in range(3, max, 2):
        if n == 3: 
            pass
        if not sieve[n]:
            for m in range(n**2, max+1, 2*n):
                sieve[m] = 1
    stop = perf_counter()
    print(f'Took {(int(stop-start)//60)}m {(stop-start)%60:0.2f}s to create sieve.')
    for n in range(max+1, 2, -1):
        if not sieve[n]:
            yield n
    
```

```python
result = 2
for prime in primes_from_largest_down(1000000000):
    
    string_list = [int(i) for i in str(prime)]
    if pandigital_test(string_list):
        result = prime
        print(f'{prime} is the largest.')
        break
```
