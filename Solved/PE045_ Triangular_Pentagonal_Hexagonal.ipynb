Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:<br>
Triangle $T_n=n(n+1)/2$     $1, 3, 6, 10, 15, \dots$<br>
Pentagonal $P_n=n(3n - 1)/2$     $1, 5, 12, 22, 35, \dots$<br>
Hexagonal $H_n=n(2n - 1)$     $1, 6, 15, 28, 45, \dots$<br>

It can be verified that T<sub>285</sub> = P<sub>165</sub> = H<sub>143</sub> = 40755.<br>
Find the next triangle number that is also pentagonal and hexagonal.<br>


```python
import math
```

```python
def pentagon_reverse(pent: int) -> float:
    return (1 + math.sqrt(1 + 24 * pent)) / 6

def hexagon_reverse(hexa: int) -> float:
    return (1 + math.sqrt(1 + 8 * hexa)) / 4

def triangle_number(number: int) -> int:
    return int(number*0.5*(number + 1))
```

```python
n = 286

while True:
    tri = triangle_number(n)
    pent = pentagon_reverse(tri)
    hexa = hexagon_reverse(tri)

    if pent%1 == 0 and hexa%1 == 0:
        break
    
    n += 1

print(f'T_{n} = {tri} is also pentagonal (index {int(pent)}) and hexagonal (index {int(hexa)})')
```
