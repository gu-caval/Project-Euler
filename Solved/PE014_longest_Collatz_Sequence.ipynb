```python
lista = {1: 1}
def countCollatz (n):
    if n in lista:
        return lista[n]
    elif n % 2 == 0:
        new = countCollatz(n/2) + 1
        lista[n] = new
        return new
    else:
        new = countCollatz(3*n+1) + 1
        lista[n] = new
        return new 
```

```python
starting = 1
longest = 0

for i in range(500000, int(1e6)):
    if i % 500 == 0: 
        print(i)
    if (size:=countCollatz(i)) > longest:
        starting = i
        longest = size

print(f"{starting=} / size={longest}")
```
